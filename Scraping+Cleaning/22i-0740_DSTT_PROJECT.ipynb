# Advanced Mobile Product Analysis and Interactive Chatbot System

```python
#Importing Neccesasry libraries 
import re
import time
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.firefox.service import Service as FirefoxService
from selenium.common.exceptions import NoSuchElementException, StaleElementReferenceException
from webdriver_manager.firefox import GeckoDriverManager
from selenium.common.exceptions import NoSuchElementException, TimeoutException
```

```python
# Set up the WebDriver
driver = webdriver.Firefox(service=FirefoxService(GeckoDriverManager().install()))
```

```python
url = 'https://www.daraz.pk/catalog/?q=Redmi+Note+12+smartphone&_keyori=ss&from=input&spm=a2a0e.home.search.go.35e34076QwJddX'
driver.get(url)



mobiles_list = []

wait_timeout=10;

max_iterations=36;


page = driver.find_element(By.XPATH, '//*[@id="root"]/div/div[3]/div/div/div[1]/div[2]')
allMobiles = page.find_elements(By.CLASS_NAME, "gridItem--Yd0sa")

for i in range(min(max_iterations, len(allMobiles))):
    
    
    mobile=allMobiles[i]
    
    href_value = mobile.find_element(By.TAG_NAME, "a").get_attribute("href")

    driver.get(href_value)
    
    # single mobile page
    for _ in range(10):
        driver.execute_script("window.scrollBy(0, 100);")
        time.sleep(0.1)
        
    
    element_present = EC.presence_of_element_located((By.CLASS_NAME, 'review-header'))

    try:
        WebDriverWait(driver, wait_timeout).until(element_present)
    except TimeoutException:
        print(f"Timed out waiting for 'review-header' element to be present.")

    
    try:
        reviewsEl=driver.find_elements(By.CLASS_NAME,'review-content-sl');
        reviews = [element.text for element in reviewsEl]
    except NoSuchElementException:
        reviews=[]

    
    time.sleep(2)
    
    points = 0;
    try:
        title = driver.find_element(By.XPATH, '//*[@id="module_product_title_1"]/div/div/span').text
    except NoSuchElementException:
        title = None
        
        
    try:
        rating = float(driver.find_element(By.XPATH, '//*[@id="module_product_review"]/div/div/div[2]/div[1]/div[1]/div[1]/span[1]').text)
    except NoSuchElementException:
        rating = 0
        
    try:
        delivery = driver.find_element(By.XPATH, '//*[@id="module_seller_delivery"]/div/div/div[3]/div/div[1]/div/div[1]/div[2]').text
        
        if(delivery=='Free'):
            points += 1000
        
    except NoSuchElementException:
        delivery = None
        
    try:
        dMall = driver.find_element(By.CLASS_NAME, 'pdp-mod-product-badge')
        dMall = 'Yes'
        points += 5000 
    except NoSuchElementException:
        dMall = 'No'
        
    try:
        seller_rating = driver.find_element(By.XPATH, '//*[@id="module_seller_info"]/div/div[2]/div[1]/div[2]')
        seller_rating = seller_rating.text.strip()
        seller_rating = int(re.sub(r'[^0-9]', '', seller_rating))
        
    except NoSuchElementException:
        seller_ratiog = 0
        
    try:
        ship_score = driver.find_element(By.XPATH, '//*[@id="module_seller_info"]/div/div[2]/div[2]/div[2]')
        ship_score = ship_score.text.strip()
        ship_score = int(re.sub(r'[^0-9]', '', ship_score))
        
    except NoSuchElementException:
        ship_score = 0
        
        
    try:
        price_element = driver.find_element(By.XPATH, '//*[@id="module_product_price_1"]/div/div/span')
        price_text = price_element.text.strip()
    
        price = int(re.sub(r'[^0-9]', '', price_text))
        
        
        points += price 
        points += 10000 * ((0.4 * (rating / 5)) + (0.4 * (seller_rating / 100)) + (0.2 * (ship_score / 100)))
        
        
    except NoSuchElementException:
        price = 0

        
    mobiles_list.append({
        "Title": title,
        "Rating": rating,
        "Shipping": delivery,
        "Daraz Mall": dMall,
        "Seller Rating": seller_rating,
        "Ship on Time score": ship_score,
        "Price":price,
        "Points":points,
        "Reviews":reviews
    })
   


    driver.back()
    
    page = driver.find_element(By.XPATH, '//*[@id="root"]/div/div[3]/div/div/div[1]/div[2]')
    allMobiles = page.find_elements(By.CLASS_NAME, "gridItem--Yd0sa")
    
driver.quit()
```

```python
df = pd.DataFrame(mobiles_list)
```

```python
df.to_csv('SET_DATA1.csv', index=False)
```
